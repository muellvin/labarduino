```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
plt.rcParams.update({'font.size': 13})



file_path = 'heating4_83ohm.txt'

# Read temperature from file
with open(file_path, 'r') as file:
    temperature = [float(line.strip()) for line in file]


def exponential_func(x, a, b, c, d):
    return a + b * np.exp(-c * (x-d))

# Generate x values (indices)
x_data = np.arange(8, len(temperature))
y_data = np.array(temperature[8:len(temperature)])

# Fit the exponential function to the data
p0 = [30, -10,0.1,8]
params, covariance = curve_fit(exponential_func, x_data, y_data, p0=p0)

# Extract the fitting parameters
a_fit, b_fit, c_fit, d_fit = params

# Generate y values based on the fitted exponential curve
y_fit = exponential_func(x_data, a_fit, b_fit, c_fit, d_fit)

# Plot the original data and the fitted curve
#plt.plot(x_data, y_data, label='Original Data', marker='o')
plt.plot(x_data, y_fit, label=f'Fit: time constant ={(1/c_fit):.2f}s', linestyle='--')
#plt.plot(x_data, exponential_func(x_data, p0[0],p0[1],p0[2],p0[3]), label=f'Fitted Exponential: a={a_fit:.2f}, b={b_fit:.2f}', linestyle=':')


plt.plot(temperature, label='Measured Data')
plt.xlabel('time [s]')
plt.ylabel('T [°C]')
plt.title('Heating with 83 Ohm resistor')
plt.grid(True)
plt.legend()
plt.savefig('Heating.pdf')

plt.show()


from scipy.stats import chisquare


y_fit_chi = y_fit * (np.array(temperature[8:]).sum() / y_fit.sum())

# Perform the Chi-Square test
chi2_stat, p_value = chisquare(np.array(temperature[8:]), f_exp=y_fit_chi)

# Output the results
print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-value: {p_value:.2f}")

```
